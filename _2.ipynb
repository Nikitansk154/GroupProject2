```python
# Удалим неважные нам признаки
train_df = train_df.drop(['RoomService','FoodCourt' ,'ShoppingMall','Spa','VRDeck','Name','Cabin' ], axis = 1)
train_df.head()
```

```python
train_df.isna().sum()
```

```python
# Заполним пропуски столбца Age медианой
train_df['Age'].fillna(train_df['Age'].median(),inplace =True)
train_df.head()
```

```python
# Пропуски в признаках HomePlanet, CryoSleep, Destination, VIP заполним модой
train_df['HomePlanet'].fillna(train_df['HomePlanet'].mode()[0],inplace =True)
train_df['CryoSleep'].fillna(train_df['CryoSleep'].mode()[0],inplace =True)
train_df['Destination'].fillna(train_df['Destination'].mode()[0],inplace =True)
train_df['VIP'].fillna(train_df['VIP'].mode()[0],inplace =True)
```

```python
train_df.isnull().sum()
```

```python
# Закодируем категориальные признаки
train_df['VIP'].replace({True:1,False:0},inplace=True)
train_df['CryoSleep'].replace({True:1,False:0},inplace=True)
train_df['HomePlanet'].replace({'Europa':0,'Earth':1,'Mars':2},inplace=True)
train_df['Destination'].replace({'TRAPPIST-1e':0,'PSO J318.5-22':1,'55 Cancri e':2},inplace=True)
```

```python
train_df.head()
```

```python
# Удаляем таргет
x_train=train_df.drop('Transported',axis=1)
y_train=train_df['Transported']
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# Так же чистим и наш тестовый датасет
test_df.drop(['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','Name', 'Cabin'],axis=1,inplace=True)
test_df.head()
```

```python
test_df.shape
```

```python
# Пропуски в признаках HomePlanet, CryoSleep, Destination, VIP заполним модой
test_df['HomePlanet'].fillna(test_df['HomePlanet'].mode()[0],inplace =True)
test_df['CryoSleep'].fillna(test_df['CryoSleep'].mode()[0],inplace =True)
test_df['Destination'].fillna(test_df['Destination'].mode()[0],inplace =True)
test_df['VIP'].fillna(test_df['VIP'].mode()[0],inplace =True)
```

```python
# Заполним пропуски столбца Age медианой
test_df['Age'].fillna(test_df['Age'].median(),inplace =True)
test_df.head()
```

```python
test_df.isnull().sum()
```

```python
# Закодируем категориальные признаки
test_df['VIP'].replace({True:1,False:0},inplace=True)
test_df['CryoSleep'].replace({True:1,False:0},inplace=True)
test_df['HomePlanet'].replace({'Europa':0,'Earth':1,'Mars':2},inplace=True)
test_df['Destination'].replace({'TRAPPIST-1e':0,'PSO J318.5-22':1,'55 Cancri e':2},inplace=True)
test_df.head()
```

```python
x_test=test_df.iloc[:,0:]
x_test
```

```python
# Нормализация данных
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
# Импортируем логистическую регирессию
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
# Загрузим метод опорных векторов
from sklearn.svm import SVC
model=SVC(kernel='rbf',random_state=0)
model.fit(x_train,y_train)
```

```python
y_test=model.predict(x_test)
y_test
```

```python
# Случайный лес
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=9, random_state=0)
rf.fit(x_train,y_train)
rf_y_pred = rf.predict(x_test)
rf_y_pred
```

```python
rf_y_pred.shape
```

```python
# Собираем данные в нужный вид
data=pd.DataFrame(rf_y_pred,columns=['Transported'])
data1=pd.DataFrame(test_df.PassengerId)
result=pd.concat([data1,data],axis=1)
result
```

```python
# Форматируем в csv
result.to_csv('Submission.csv',index=False)
```
